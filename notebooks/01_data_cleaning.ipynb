# Data cleaning -
### [Spotify Tracks Dataset] 



```python
import pandas as pd 
# load data 
df = pd.read_csv("../data/raw/spotify-tracks-dataset-detailed.csv") 
# dimensions
df.shape  # (114000, 20)
```

### First look 

```python
df.info()
df.head()
```

### [ checking data quality ]

```python
#nulls + duplicates

df.isnull().sum()
df[df["artists"].isnull()] # only one raw has  null vlaues for [artists ,album_name,	track_name]

df.duplicated().sum() # 450 values are duplicated in data 
data = df[df.duplicated()]
data 
```

```python
# handle missing values( only one raw(65900) has missing value for [artists ,album_name, track_name] thats why we drop the raw )
df = df[df["track_id"] != "1kR4gIb7nGxHPI3D2ifs59"] 
```

```python
# handle duplicates [there are total |450| duplicates in our data but we cant take them with us :(  so ]
df = df.drop_duplicates()
print("After:", df.shape)
```

```python
# checking for data type 
df.dtypes
# All datatypes are fine but lets convert the duration_ms into [duration_minuts ]
```

```python
# fix dtypes
df["duration_min"] = df["duration_ms"]/60000
# new column added minutes will be usfull in EDA
```

```python
# Check audiio features impossible values
print(df[["popularity", "danceability", "energy", "duration_min"]].describe())
# seems good 
```

## Summary of cleaning steps

- Dropped **1 row(65900th)** with missing three values of[artists ,album_name, track_name]
- Removed **450** duplicate records  from the dataset 
- added an new column by (Converted duration to minutes)
- Checked for **invalid or impossible values** in the audio feature columns; no unusual values were found
- final shape :  (113549, 20)
- The dataset is now clean and ready for Exploratory Data Analysis (EDA)
