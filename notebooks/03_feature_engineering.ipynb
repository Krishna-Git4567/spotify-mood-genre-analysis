# Feature Engineering

```python
import pandas as pd 
```

```python
df = pd.read_csv("../data/processed/clean_tracks.csv")
```

### New feature : mode_score

```python
df["mood_score"] = (df["valence"] + df["energy"] )/2
df[["mood_score"]]
```

### New feature : party_score

```python
df["party_score"] = (df["danceability"] + df["energy"]) /2
df[["party_score"]]
```

### New feature: popularity_tier

```python
df["popularity_tier"] =  pd.cut(    
    df["popularity"],
    bins= [0, 30 , 60 ,100], 
    labels= ["Low", "Medium", "High"],
    include_lowest= True  )

df[["popularity","popularity_tier"]].sample(10)
```

### New feature : tempo_levels

```python
df["tempo_level"] = pd.cut(
    df["tempo"],
    bins= [0, 90, 130 , 250 ] ,
    labels= ["Slow", "Medium", "Fast"],
    include_lowest = True
)

df[["tempo", "tempo_level"]].sample(10)
```

```python
df.to_csv('../data/processed/featured_tracks.csv', index=False)
```

## SUMMARY >
- Created mood_score by combining features related to a song's emotional characteristics
- Created party_score byp combining [danceability and energy] to represent how suitable a song is for dancing or parties
- Created popularity_tier by categorizing the continuous popularity score into [Low, Medium, and High] groups. This can be useful for  classification tasks
- Created tempo_level by grouping the song's tempo (measured in beats per minute, BPM) into categories such as [Slow, Medium, and Fast]
